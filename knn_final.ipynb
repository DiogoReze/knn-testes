```python
# 1 -> pronado
# 2 -> neutro
# 3 -> supinado

pisadaE = []

for i in range(len(s1eNormal)):
  if (s2eNormal[i] > s3eNormal[i]) and (s8eNormal[i] > s6eNormal[i]):
    pisadaE.append(3)
  elif (s2eNormal[i] < s3eNormal[i]) and (s8eNormal[i] < s6eNormal[i]):
    pisadaE.append(1)
  else:
    pisadaE.append(2)

pisadaD = []

for i in range(len(s1dNormal)):
  if (s2dNormal[i] > s3dNormal[i]) and (s8dNormal[i] > s6dNormal[i]):
    pisadaD.append(3)
  elif (s2dNormal[i] < s3dNormal[i]) and (s8dNormal[i] < s6dNormal[i]):
    pisadaD.append(1)
  else:
    pisadaD.append(2)

pisadaE = np.array(pisadaE)
s1eNormal = np.array(s1eNormal); s2eNormal = np.array(s2eNormal); s3eNormal = np.array(s3eNormal); 
s4eNormal = np.array(s4eNormal); s5eNormal = np.array(s5eNormal); s6eNormal = np.array(s6eNormal); 
s7eNormal = np.array(s7eNormal); s8eNormal = np.array(s8eNormal); s9eNormal = np.array(s9eNormal); 

#print((s1eNormal))

matrizT = np.array([s1eNormal.T, s2eNormal.T, s3eNormal.T, s4eNormal.T, s5eNormal.T, s6eNormal.T, s7eNormal.T, s8eNormal.T, s9eNormal.T, pisadaE])
matrizT = np.transpose(matrizT)
print(matrizT.shape)

matrizTT = np.array([s1eNormal.T, s2eNormal.T, s3eNormal.T, s4eNormal.T, s5eNormal.T, s6eNormal.T, s7eNormal.T, s8eNormal.T, s9eNormal.T])
matrizTT = np.transpose(matrizTT)
pisadaE = np.transpose(pisadaE)
print(pisadaE)

```

```python
# dividir dados teste e dados treinamento
# https://estatsite.com.br/2020/07/25/dividindo-o-dataset-em-treino-e-teste-no-python/
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(matrizTT, pisadaE, train_size=0.85)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_step = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p = 2)
knn_step.fit(X_train, y_train)
```

```python
previsoes = knn_step.predict(X_test)
previsoes
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(knn_step)
cm.fit(X_train, y_train)
cm.score(X_test, y_test)
```
