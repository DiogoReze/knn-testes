```python
from random import randint
import matplotlib.pyplot as plt

#gerar um número aleatorio dentro dos codigos de cargos (1-64)

n_cargo_amostra = randint(1,64)

#com base no cargo e seu salários de 2017-2022, mostrar o cargo
#e a evolução do salário, juntamente com o salário real já deduzido a inflação

cargo_amostra = cargos[n_cargo_amostra]
s = [salario2017[n_cargo_amostra], salario2018[n_cargo_amostra], salario2019[n_cargo_amostra], salario2020[n_cargo_amostra], salario2021[n_cargo_amostra], salario2022[n_cargo_amostra]]
ano = [2017, 2018, 2019, 2020, 2021, 2022]
inflacao = [2.95, 3.75, 4.31, 10.74, 10.06, 8.8]

s_real = []
#print(inflacao)

for i in range(len(ano)):
  s_temp = s[i] * (1 - inflacao[i]/100.0)
  s_real.append(s[i] - (s[i] - s_temp))

plt.figure(figsize=(18,6)) 
plt.plot(ano, s, lw='2', label = 'Salário Reajustado', color = 'blue')
plt.scatter(ano, s, color = 'blue')
plt.plot(ano, s_real, lw='2', label = 'Salário real', color = 'red')
plt.scatter(ano, s_real, color = 'red')
plt.legend(fontsize=20)
plt.xlabel('Anos', fontsize=20)
plt.ylabel('Salário', fontsize=20)
for i in range(len(s)):
    plt.annotate("R$ {:.2f}".format(s[i]), (ano[i], s[i]), fontsize=16)
    plt.annotate("R$ {:.2f}".format(round(s_real[i],2)), (ano[i], s_real[i]), fontsize=16)
plt.title(cargo_amostra, fontsize=20)
plt.show()
```

```python
#classificando cada cargo com relação a classe social

salarioMin17 = 937.00; salarioMin18 = 994.00; salarioMin19 = 998.00; 
salarioMin20 = 1045.00; salarioMin21 = 1100.00; salarioMin22 = 1212.00;

classeSocial_17 = []
classeSocial_18 = []
classeSocial_19 = []
classeSocial_20 = []
classeSocial_21 = []
classeSocial_22 = []

#classificação da classe social em 2017
for i in range(len(cargos)):
  if salario2017[i] <= 2 * salarioMin17:
    classeSocial_17.append('E')
  elif (salario2017[i] > (2 * salarioMin17)) and (salario2017[i] <= (4 * salarioMin17)):
    classeSocial_17.append('D')
  else:
    classeSocial_17.append('C')

#classificação da classe social em 2018
for i in range(len(cargos)):
  if salario2018[i] <= 2 * salarioMin18:
    classeSocial_18.append('E')
  elif (salario2018[i] > (2 * salarioMin18)) and (salario2018[i] <= (4 * salarioMin18)):
    classeSocial_18.append('D')
  else:
    classeSocial_18.append('C')

#classificação da classe social em 2019
for i in range(len(cargos)):
  if salario2019[i] <= 2 * salarioMin19:
    classeSocial_19.append('E')
  elif (salario2019[i] > (2 * salarioMin19)) and (salario2019[i] <= (4 * salarioMin19)):
    classeSocial_19.append('D')
  else:
    classeSocial_19.append('C')

#classificação da classe social em 2020
for i in range(len(cargos)):
  if salario2020[i] <= 2 * salarioMin20:
    classeSocial_20.append('E')
  elif (salario2020[i] > (2 * salarioMin20)) and (salario2020[i] <= (4 * salarioMin20)):
    classeSocial_20.append('D')
  else:
    classeSocial_20.append('C')

#classificação da classe social em 2021
for i in range(len(cargos)):
  if salario2021[i] <= 2 * salarioMin21:
    classeSocial_21.append('E')
  elif (salario2021[i] > (2 * salarioMin21)) and (salario2021[i] <= (4 * salarioMin21)):
    classeSocial_21.append('D')
  else:
    classeSocial_21.append('C')

#classificação da classe social em 2022
for i in range(len(cargos)):
  if salario2022[i] <= 2 * salarioMin22:
    classeSocial_22.append('E')
  elif (salario2022[i] > (2 * salarioMin22)) and (salario2022[i] <= (4 * salarioMin22)):
    classeSocial_22.append('D')
  else:
    classeSocial_22.append('C')
```

```python
#sexo do trabalhador
from random import randrange, uniform
genero = []
contM = 0
contF = 0

for i in range(len(cargos)):
  n = uniform(1, 10)
  #print(n)
  if uniform(1, 10) <= 9:
    genero.append('Masculino')
    contM = contM + 1
  else:
    genero.append('Feminino')
    contF = contF + 1

print(contF,contM)
print(100*contF/len(cargos),100*contM/len(cargos))
```

```python
# escolaridade
# t - curso técnico, curso inerente a função
# s - ensino superior, curso no cargo
escol = []
contT = 0
contS = 0

for i in range(len(cargos)):
  if  uniform(1, 20) <= 17:
    escol.append('técnico inerente a função')
    contT = contT + 1
  else:
    escol.append('superior e curso para o respectivo cargo')
    contS = contS + 1
    
print(escol)
print(100*contS/len(cargos),100*contT/len(cargos))
```
